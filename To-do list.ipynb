```python
#codsoft task to make to-do list
#tkinter is being used
from tkinter import *
import tkinter as tk
from tkinter.font import Font
from tkinter import filedialog
import pickle


#creation of window
root = tk.Tk()
root.title("To-do List")
root.geometry("550x400")

#Defining Font style
font = Font( 
    family = "Times New Roman",
    size = 14,
    weight = "bold",)

#creation of frame
frame = Frame(root)
frame.pack(pady = 20)

#creation of list box
list_box = Listbox(frame,
                   font = font,
                   width = 45,
                   height = 5,
                   bg = "SystemButtonFace",
                   bd = 0,
                   fg = "#201E43",
                   highlightthickness = 0,
                   selectbackground = "#a6a6a6",
                   activestyle ="none")

list_box.pack(side = LEFT , fill = BOTH)

#creation of test list
#items = ["take nap","make presentation","guitar lessons","learn korean"]

#adding test list to listbox
#for i in items:
 #   list_box.insert(END,i)

#creation of scrollbar
scroll_bar = Scrollbar(frame)
scroll_bar.pack(side = RIGHT , fill = BOTH)

#add scrollbar
list_box.config(yscrollcommand = scroll_bar.set)
scroll_bar.config(command = list_box.yview)

#creating button frame
button_frame = Frame(root)
button_frame.pack(pady = 20)


#entry box to add items to list
input_space = Entry(root,font = ("Helvetica", 16), width = 43) 
input_space.pack(pady = 20)



#creation of button events
def delete_item():
    list_box.delete(ANCHOR)

def add_item():
    list_box.insert(END,input_space.get())
    input_space.delete(0,END)
    
def complete_item():
    #to show completed items
    list_box.itemconfig(
        list_box.curselection(),
        fg = "#387F39"
    )
    #removing selection bar
    list_box.selection_clear(0,END)
    
def uncompleted_item():
    #to show uncompleted items
    list_box.itemconfig(
        list_box.curselection(),
        fg = "#201E43"
    )
    #removing selection bar
    list_box.selection_clear(0,END)

def remove():
    #to delete all crossed_off item
    count = 0
    while count < list_box.size():
        if list_box.itemcget(count,"fg") == "#387F39" :
            list_box.delete(list_box.index(count))

        else:
            count += 1
    

#creation of drop down menu
me_nu = Menu(root)
root.config(menu=me_nu)

#creation of functions for dropdown menu items
def save_list():
    file_name = filedialog.asksaveasfilename(
        initialdir = "C:codsoft_internship_projects/coding stuff",
        title = "Save File",
        filetypes = (("Dat Files","*.dat"),
                     ("All Files","*.*"))
    )
    if file_name:
        if file_name.endswith(".dat"):
            pass
        else:
           file_name = f'{file_name}.dat'
            
        #to delete all crossed_off item
        count = 0
        while count < list_box.size():
          if list_box.itemcget(count,"fg") == "#387F39" :
             list_box.delete(list_box.index(count))

          else:
             count += 1
        #grab all items from list
        items = list_box.get(0,END)
       #open selected file 
        output_file = open(file_name,'wb')
       #add items from list to file
        pickle.dump(items,output_file)
    
    
def open_list():
    file_name = filedialog.askopenfilename(
        initialdir = "C:codsoft_internship_projects/coding stuff",
        title = "Open File",
        filetypes = (("Dat Files","*.dat"),
                     ("All Files","*.*"))
    )
    if file_name:
        #delete currently open list
        list_box.delete(0,END)
        
        #open  the file
        input_file = open(file_name,'rb')
        
        #load the info from file
        items = pickle.load(input_file)
        #output info across screen
        for i in  items:
            list_box.insert(END,i)
                                            

def clear_list():
    list_box.delete(0,END)


#items in menu
file_menu = Menu(me_nu,tearoff = False)
me_nu.add_cascade(label = "File", menu = file_menu)

#addition of dropdown items
file_menu.add_command(label = "Save", command = save_list)
file_menu.add_separator()
file_menu.add_command(label = "Open", command = open_list)
file_menu.add_separator()
file_menu.add_command(label = "Clear", command = clear_list)




                  

#Adding buttons
delete_button = Button(button_frame, text = "DELETE ITEM",command = delete_item )
add_button = Button(button_frame,text = "ADD ITEM",command = add_item )
completed_button = Button(button_frame, text = "COMPLETED",command = complete_item )
uncompleted_button = Button(button_frame,text = "UNCOMPLETED",command = uncompleted_item)
remove_completed_item_button = Button(button_frame,text = "DELETE COMPLETED ",command = remove)



#positioning buttons
delete_button.grid( row = 0 ,column = 0 )
add_button.grid(row = 0 ,column = 1 , padx = 20  )
completed_button.grid(row = 0 ,column = 2 )
uncompleted_button.grid(row = 0 ,column = 3, padx = 20)
remove_completed_item_button.grid(row = 0,column = 4)





                    



root.mainloop()


```
