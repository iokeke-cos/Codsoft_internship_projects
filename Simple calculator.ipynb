```python
#Project to make simple calculator using tkinter
import tkinter as tk
from tkinter import *

#creation of window
root = tk.Tk()
root.title ("Simple Calculator Project") 
root.config(bg="black")
root.geometry("504x520")

#creation of button events
def button_event(digit):
    entry = input_space.get()
    input_space.delete(0,END)
    input_space.insert(0, str(entry) +  str(digit)) 
    

def button_clear():
    input_space.delete(0,END)

def button_add():
    first_digit = input_space.get()
    global f_dig
    global operations
    operations = "add"
    f_dig =int(first_digit)
    input_space.delete(0,END)
    
    
def button_subtract():
    first_digit = input_space.get()
    global f_dig
    global operations
    operations = "minus"
    f_dig =int(first_digit)
    input_space.delete(0,END)
    

def button_divide():
    first_digit = input_space.get()
    global f_dig
    global operations
    operations = "divide"
    f_dig =int(first_digit)
    input_space.delete(0,END)
    


def button_multiply():
    first_digit = input_space.get()
    global f_dig
    global operations
    operations = "multiply"
    f_dig =int(first_digit)
    input_space.delete(0,END)
    


def button_delete():
    delete_len = len(input_space.get())
    input_space.delete(delete_len - 1,END)
    
     
def button_equals():
    second_digit = input_space.get()
    s_dig = int(second_digit)
    if operations == "add":
        input_space.delete(0,END)
        input_space.insert(0,f_dig + s_dig)

    if operations == "minus":
        input_space.delete(0,END)
        input_space.insert(0,f_dig - s_dig)  

    if operations == "multiply":
        input_space.delete(0,END)
        input_space.insert(0,f_dig * s_dig)

    if operations == "divide":
        input_space.delete(0,END)
        input_space.insert(0,f_dig / s_dig)
  

    
     
    
#creation of buttons
button_1 = tk.Button(root,text = "1",padx = 70,pady = 20, command= lambda: button_event(1))
button_2 = tk.Button(root,text = "2",padx = 72,pady = 20, command= lambda: button_event(2))
button_3 = tk.Button(root,text = "3",padx = 73,pady = 20, command= lambda: button_event(3))
button_4 = tk.Button(root,text = "4",padx = 70,pady = 20, command= lambda: button_event(4))
button_5 = tk.Button(root,text = "5",padx = 72,pady = 20, command= lambda: button_event(5))

button_6 = tk.Button(root,text = "6",padx = 73,pady = 20, command= lambda: button_event(6))
button_7 = tk.Button(root,text = "7",padx = 70,pady = 20, command= lambda: button_event(7))
button_8 = tk.Button(root,text = "8",padx = 72,pady = 20, command= lambda: button_event(8))
button_9 = tk.Button(root,text = "9",padx = 73,pady = 20, command= lambda: button_event(9))
button_0 = tk.Button(root,text = "0",padx = 72,pady = 20, command= lambda: button_event(0))


clear_button = tk.Button(root,text="clear",padx = 60, pady = 20,command = button_clear)
delete_button = tk.Button(root,text = "delete",padx = 52,pady = 20,command = button_delete)
plus_button = tk.Button(root,text = "+" , padx = 71, pady = 23,command = button_add)
minus_button = tk.Button(root,text = "-",padx = 73, pady = 23,command = button_subtract)
multiply_button = tk.Button(root,text = "x",padx = 74, pady = 23,command = button_multiply)
divide_button = tk.Button(root,text = "/",padx = 72, pady = 26,command = button_divide)
equal_button = tk.Button(root,text = "=",padx = 157 , pady = 26,command =button_equals)



#button placements
button_1.grid(row =3,column=0)
button_2.grid(row =3,column=1)
button_3.grid(row =3,column=2)
button_4.grid(row =2,column=0)
button_5.grid(row =2,column=1)

button_6.grid(row =2,column=2)
button_7.grid(row =1,column=0)
button_8.grid(row =1,column=1)
button_9.grid(row =1,column=2)
button_0.grid(row =4,column=1)

delete_button.grid(row = 4,column=0)
clear_button.grid(row = 4,column = 2)
plus_button.grid(row=5,column=0)
minus_button.grid(row=5,column=1)
multiply_button.grid(row=5,column=2)
divide_button.grid(row=6,column=0)
equal_button.grid(row=6,column= 1,columnspan=2,padx=5,pady=5)



#styling buttons
button_1.config(font ="8", fg = "white",bg = "#03346E")
button_2.config(font ="8",fg = "white",bg = "#03346E")
button_3.config(font ="8",fg = "white",bg = "#03346E")
button_4.config(font ="8",fg = "white",bg = "#03346E")
button_5.config(font ="8",fg = "white",bg = "#03346E")

button_6.config(font ="8",fg = "white",bg = "#03346E")
button_7.config(font ="8",fg = "white",bg = "#03346E")
button_8.config(font ="8",fg = "white",bg = "#03346E")
button_9.config(font ="8",fg = "white",bg = "#03346E")
button_0.config(font ="8",fg = "white",bg = "#03346E")

clear_button.config(font ="4",fg = "white",bg = "#03346E")
plus_button.config(font = "18",fg = "white",bg = "#2387BA")
minus_button.config(font = "18",fg = "white",bg = "#2387BA")
multiply_button.config(font = "18",fg = "white",bg = "#2387BA")
divide_button.config(font = "18",fg = "white",bg = "#2387BA")
delete_button.config(font ="4",fg ="white",bg = "#03346E")
equal_button.config(font = "18",fg ="white",bg = "#03346E")
                    


#creation of input/entry space

input_space = tk.Entry(root, width = 70, borderwidth =10)
input_space.grid(row =0,column = 0,columnspan=3,padx = 20,  pady = 20 )


root.mainloop()
```
